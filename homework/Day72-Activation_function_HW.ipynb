```python
import numpy as np
import matplotlib.pylab as plt
```

```python
def sigmoid(x):
    return (1/(1+np.exp(-x)))
```

```python
x = plt.linspace(-10,10,100)
plt.plot(x, sigmoid(x), 'b', label='linspace(-10,10,10)')
```

```python
def softmax(x):
    return np.exp(x)/sum(np.exp(x))
```

```python
plt.plot(x,softmax(x))
```

```python
def ReLU(x):
    return abs(x)*(x>0)
```

```python
plt.plot(x,ReLU(x))
```

```python
def dReLU(x):
    return 1*(x>0)
```

```python
plt.plot(x,dReLU(x))
```
